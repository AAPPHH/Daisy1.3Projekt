```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"game_history.xlsx"
data = pd.read_excel(file_path)
```

```python
data['winner'] = pd.Categorical(data['winner'])

# Plotten
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='N_turn', y='winner', palette='deep')

plt.title('Scatter Plot of Game Winner vs Number of Turns')
plt.xlabel('Number of Turns (N_turn)')
plt.ylabel('Winner')

plt.show()
```
